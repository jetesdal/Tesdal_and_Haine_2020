```python
def get_gdf_patch(coords_patch):
    # Replace each pair of points by 10 points for projected map
    geometry = []
    segments = zip(coords_patch[:-2:2], coords_patch[1:-1:2], coords_patch[2::2], 
                   coords_patch[3::2])
    
    for line in [shapely.geometry.LineString([(x1, y1), (x2, y2)]) for (x1,y1,x2,y2) in segments]:
        for length in np.linspace(0,1,11):
            geometry.append( line.interpolate(length, normalized=True) )
        
    gdf_patch = gpd.GeoDataFrame([], geometry=geometry)
    
    # Convert Points to Polygo
    gdf_patch['geometry'] = gdf_patch['geometry'].apply(lambda x: x.coords[0])
    
    gdf_patch['shape_id'] = 0
    gdf_patch = gdf_patch.groupby('shape_id')['geometry'].apply(lambda x: shapely.geometry.Polygon(x.tolist())).reset_index()
    gdf_patch = gpd.GeoDataFrame(gdf_patch, geometry = 'geometry')
    
    # Salem uses this attribute:
    gdf_patch.crs = {'init': 'epsg:4326'}
    
    return gdf_patch
```

### Heat anomaly budget

```python
# Density kg/m^3
rhoconst = 1029

# Heat capacity (J/kg/K)
c_p = 3994

# Watts to Petawatts
pw = 1e-15
```

```python
terms = ['tendH', 'forcH', 'adv_ConvH', 'adv_Uclim_Tanom', 'adv_Uanom_Tclim', 'adv_Uanom_Tanom', 
         'dif_ConvH','resH','totalH']

budgH = []

for mask in [main.mask_spna,main.mask_nsea,main.mask_lsea]:
    
    # Temperature tendency (C m^3/s)
    tendH = (budgH.tnd*mask.where(mask==1)).sum(dim=['tile','k','j','i'])
    
    # Temperature forcing (C m^3/s)
    forcH = (budgH.frc*mask.where(mask==1)).sum(dim=['tile','k','j','i'])
    
    # Convergence of mean advection of anomaleous temperature (C m^3/s)
    adv_Uclim_Tanom = ((budgH.hadv_Uclim_Tanom\
                        +budgH.vadv_Uclim_Tanom)*mask.where(mask==1)).sum(dim=['tile','k','j','i'])
    
    # Convergence of anomaleous advection of mean temperature (C m^3/s)
    adv_Uanom_Tclim = ((budgH.hadv_Uanom_Tclim\
                         +budgH.vadv_Uanom_Tclim)*mask.where(mask==1)).sum(dim=['tile','k','j','i'])
    
    # Convergence of nonlinear temperature advection (C m^3/s)
    adv_Uanom_Tanom = ((budgH.hadv_Uanom_Tanom\
                         +budgH.vadv_Uanom_Tanom)*mask.where(mask==1)).sum(dim=['tile','k','j','i'])
    
    # Diffusion (C m^3/s)
    dif_ConvH = ((budgH.hdif+budgH.vdif)*mask.where(mask==1)).sum(dim=['tile','k','j','i'])
    
    # Residual
    resH = (budgH.res*mask.where(mask==1)).sum(dim=['tile','k','j','i'])
    
    # Convergence of total advection (C m^3/s)
    adv_ConvH = adv_Uclim_Tanom + adv_Uanom_Tclim + adv_Uanom_Tanom 
    
    # Sum of terms (m^3/s)
    totalH = forcH + adv_ConvH + dif_ConvH + resH
    
    results = dask.compute(tendH.reset_coords(drop=True)*c_p*rhoconst, forcH.reset_coords(drop=True)*c_p*rhoconst, 
                           adv_ConvH.reset_coords(drop=True)*c_p*rhoconst, 
                           adv_Uclim_Tanom.reset_coords(drop=True)*c_p*rhoconst,
                           adv_Uanom_Tclim.reset_coords(drop=True)*c_p*rhoconst, 
                           adv_Uanom_Tanom.reset_coords(drop=True)*c_p*rhoconst,
                           dif_ConvH.reset_coords(drop=True)*c_p*rhoconst, resH.reset_coords(drop=True)*c_p*rhoconst,
                           totalH.reset_coords(drop=True)*c_p*rhoconst)
    
    budg = xr.concat(results,pd.Index(terms, name='term'))
    budgH.append(budg)
```

```python
region = ['spna', 'nsea', 'lsea']
budgH = xr.concat(budgH,pd.Index(region, name='region'))
```
